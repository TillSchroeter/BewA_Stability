```python
### Bibliotheken
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
### Daten einlesen

# Einlesen in einen DataFrame
df = pd.read_csv("2025-04-11-11-09_Pre_C_Li_3_Kopie.csv",delimiter=';',encoding='utf-8',decimal=',')  # ðŸ‘ˆ Wichtig: Komma als Dezimaltrennzeichen
# df = pd.read_c("Daten_Krepper_Elisabeth_D/2025-04-11-11-09_Pre_C_Li_3_Kopie.csv", delimiter=';', encoding='utf-8')  # ggf. encoding anpassen
# Erste Zeilen anzeigen
# print(df ["time"])

### Daten extrahieren
# Zeitspalte
time = df["time"]  # Zeit in Sekunden
time = time - time[0]  # Zeit relativ zur ersten Zeit
print(time)
### Gelenkswinkel-Spalten

# Gelenkwinkel â€“ linkes Bein
lt_hip_flexion = df["LT Hip Flexion (deg)"]
lt_hip_abduction = df["LT Hip Abduction (deg)"]
lt_hip_rotation_ext = df["LT Hip Rotation Ext (deg)"]
lt_knee_abduction = df["LT Knee Abduction (deg)"]
lt_knee_rotation_ext = df["LT Knee Rotation Ext (deg)"]
lt_knee_flexion = df["LT Knee Flexion (deg)"]
lt_knee_extension = df["Noraxon MyoMotion-Joints-Knee LT-Extension (deg)"]
lt_ankle_dorsiflexion = df["LT Ankle Dorsiflexion (deg)"]
lt_ankle_abduction = df["LT Ankle Abduction (deg)"]
lt_ankle_inversion = df["LT Ankle Inversion (deg)"]

# Gelenkwinkel â€“ rechtes Bein
rt_hip_flexion = df["RT Hip Flexion (deg)"]
rt_hip_abduction = df["RT Hip Abduction (deg)"]
rt_hip_rotation_ext = df["RT Hip Rotation Ext (deg)"]
rt_knee_abduction = df["RT Knee Abduction (deg)"]
rt_knee_rotation_ext = df["RT Knee Rotation Ext (deg)"]
rt_knee_flexion = df["RT Knee Flexion (deg)"]
rt_knee_extension = df["Noraxon MyoMotion-Joints-Knee RT-Extension (deg)"]
rt_ankle_dorsiflexion = df["RT Ankle Dorsiflexion (deg)"]
rt_ankle_abduction = df["RT Ankle Abduction (deg)"]
rt_ankle_inversion = df["RT Ankle Inversion (deg)"]

# Torso-Pelvis-Winkel
torso_pelvic_flexion_fwd = df["Torso-Pelvic Flexion Fwd (deg)"]

lt_torso_pelvic_flexion_lat = df["LT Torso-Pelvic Flexion Lat (deg)"]
lt_torso_pelvic_axial = df["LT Torso-Pelvic Axial (deg)"]
rt_torso_pelvic_flexion_lat = df["RT Torso-Pelvic Flexion Lat (deg)"]
rt_torso_pelvic_axial = df["RT Torso-Pelvic Axial (deg)"]

###########################################################################################################################################################
```

```python
### Plottten
# Subplots: 3 Zeilen (Hip, Knee, Ankle), 2 Spalten (LT, RT)
fig, axes = plt.subplots(4, 2, figsize=(20, 12))
fig.suptitle("Gelenkwinkel Ã¼ber der Zeit (Sprung mit linken bein)", fontsize=16)

# Hip Flexion
axes[0, 0].plot(time, lt_hip_flexion, label="LT Hip Flexion", color="blue")
axes[0, 0].set_ylabel("Hip (deg)")
axes[0, 0].legend()

axes[0, 1].plot(time, rt_hip_flexion, label="RT Hip Flexion", color="red")
axes[0, 1].set_ylabel("Hip (deg)")
axes[0, 1].legend()

# Knee Flexion
axes[1, 0].plot(time, lt_knee_flexion, label="LT Knee Flexion", color="blue")
axes[1, 0].set_ylabel("Knee (deg)")
axes[1, 0].legend()

axes[1, 1].plot(time, rt_knee_flexion, label="RT Knee Flexion", color="red")
axes[1, 1].set_ylabel("Knee (deg)")
axes[1, 1].legend()
# Ankle Dorsiflexion
axes[2, 0].plot(time, lt_ankle_dorsiflexion, label="LT Ankle Dorsiflexion", color="blue")
axes[2, 0].set_ylabel("Ankle (deg)")
axes[2, 0].set_xlabel("Time (s)")
axes[2, 0].legend()

axes[2, 1].plot(time, rt_ankle_dorsiflexion, label="RT Ankle Dorsiflexion", color="red")
axes[2, 1].set_ylabel("Ankle (deg)")
axes[2, 1].legend()

# lt_torso_pelvic_flexion_lat
axes[3, 0].plot(time, lt_torso_pelvic_flexion_lat, label="LT Torso-Pelvic Flexion Lat", color="blue")
axes[3, 0].set_ylabel("Torso-Pelvis (deg)")
axes[3, 0].set_xlabel("Time (s)")
axes[3, 0].legend()

axes[3, 1].plot(time, rt_torso_pelvic_flexion_lat, label="RT Torso-Pelvic Flexion Lat", color="red")
axes[3, 1].set_ylabel("Torso-Pelvis (deg)")
axes[3, 1].set_xlabel("Time (s)")
axes[3, 1].legend()

# plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Platz fÃ¼r den Titel lassen
plt.show()

# Pelvis Flexion
plt.figure(figsize=(10, 4))
plt.plot(time, torso_pelvic_flexion_fwd, label="Pelvis Flexion", color="green")
plt.title("Pelvis Flexion Ã¼ber der Zeit")
plt.xlabel("Time (s)")
plt.ylabel("Pelvis Flexion (deg)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
