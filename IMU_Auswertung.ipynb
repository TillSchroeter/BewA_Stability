# Erste Stabilisationsphase 
(Knie, HÃ¼ft, Sprunggelenkswinkel verÃ¤ndern sich nicht mehr viel nach landung)

```python
### Funktion zur Bestimmung des stabilen Zeitpunkts

def find_stable_time(time, signal, peak_time, slope_threshold=0.05):
    # BeschrÃ¤nke die Daten auf Zeiten nach dem Peak
    mask = time > peak_time
    time_post = time[mask].reset_index(drop=True)
    signal_post = signal[mask].reset_index(drop=True)

    # Berechne numerische Ableitung (Steigung)
    slopes = np.gradient(signal_post, time_post)

    # Finde den ersten Index, wo Ableitung unter dem Schwellenwert fÃ¤llt
    stable_indices = np.where(np.abs(slopes) < slope_threshold)[0]
    if len(stable_indices) == 0:
        # Falls kein stabiler Punkt gefunden wurde, gib NaN oder Ende zurÃ¼ck
        return np.nan
    else:
        # Erster stabiler Zeitpunkt
        stable_time = time_post.iloc[stable_indices[0]]
        return stable_time
    
keys = ("Re_1_Pre","Re_2_Pre","Re_3_Pre")  # Beispiel: Re_3_Pre
joints = ("RT Knee Flexion (deg)", "RT Hip Flexion (deg)", "RT Ankle Dorsiflexion (deg)")
treshholds = (0.25, 0.25, 0.05)  # Schwellenwerte durch probieren

# knee_flexion, time, peak_time = data[key]["df"]["RT Knee Flexion (deg)"], data[key]["df"]["time"], data[key]["landing_peak_time"]
# kip_flexion, time, peak_time = data[key]["df"]["RT Hip Flexion (deg)"], data[key]["df"]["time"], data[key]["landing_peak_time"]
# ankle_flexion, time, peak_time = data[key]["df"]["RT Ankle Dorsiflexion (deg)"], data[key]["df"]["time"], data[key]["landing_peak_time"]

# for key in keys:
#     for joint, th in zip (joints, treshholds):
#         flexion = data[key]["df"][joint]
#         time = data[key]["df"]["time"]
#         peak_time = data[key]["landing_peak_time"]
#         slope_threshold = th
#         # Stabiler Zeitpunkt finden
#         stable_time = find_stable_time(time, flexion, peak_time, slope_threshold)
#         print(f"Stabiler Zeitpunkt fÃ¼r {key} ({joint}): {stable_time:.3f} s")

# Dictionary zum Speichern der mittleren stabilen Zeitpunkte je Messung
mean_stable_times = {}

for key in keys:
    stable_times = []

    for joint, th in zip(joints, treshholds):
        flexion = data[key]["df"][joint]
        time = data[key]["df"]["time"]
        peak_time = data[key]["landing_peak_time"]
        slope_threshold = th

        stable_time = find_stable_time(time, flexion, peak_time, slope_threshold)
        stable_times.append(stable_time)

        print(f"Stabiler Zeitpunkt fÃ¼r {key} ({joint}): {stable_time:.3f} s")

    # Mittelwert der drei Zeitpunkte (sofern keine NaNs enthalten)
    stable_times = [t for t in stable_times if not np.isnan(t)]
    if stable_times:
        mean_time = np.mean(stable_times)
    else:
        mean_time = np.nan

    mean_stable_times[key] = mean_time
    print(f"âž¤ Mittlerer stabiler Zeitpunkt fÃ¼r {key}: {mean_time:.3f} s\n")


### in Datenstruktur speichern
for key in mean_stable_times:
    data[key]["stable_time"] = mean_stable_times[key]

# stable_time = find_stable_time(time, knee_flexion, peak_time, slope_threshold=0.25)
# print(f"Stabiler Zeitpunkt fÃ¼r {key}: {stable_time:.3f} s")

# Plotten
# plt.figure(figsize=(10,5))
# plt.plot(time, knee_flexion, label="Knie-Flexion")
# plt.axvline(x=peak_time, color='red', linestyle='--', label="Landing Peak")
# if not np.isnan(stable_time):
#     plt.axvline(x=stable_time, color='green', linestyle='--', label="Stabiler Zeitpunkt")
# plt.xlabel("Zeit (s)")
# plt.ylabel("Knie-Flexion (Â°)")
# plt.title(f"Knie-Flexion und stabiler Zeitpunkt fÃ¼r {key}")
# plt.legend()
# plt.grid(True)
# plt.show()
```
